```python
def entropy(df, label):
    class1_df = df[df[label] == 1]
    class2_df = df[df[label] == 2]
    class3_df = df[df[label] == 3]
    class1 = class1_df.shape[0]
    class2 = class2_df.shape[0]
    class3 = class3_df.shape[0]
    total_values = class1 + class2 + class3
    prob_class1 = (class1/total_values)
    prob_class2 = (class2/total_values)
    prob_class3 = (class3/total_values)

    #if class1 == 0 or class2 == 0 or class3 == 0:
    if (class1 == 0 and class2 == 0) or (class1 == 0 and class3 == 0) or (class2 == 0 and class3 == 0):
        entropy_split = 0
    elif class1 == 0:
        entropy_split = -((prob_class2 * math.log(prob_class2, 2)) + (prob_class3 * math.log(prob_class3, 2)))
    elif class2 == 0:
        entropy_split = -((prob_class1 * math.log(prob_class1, 2)) + (prob_class3 * math.log(prob_class3, 2)))
    elif class3 == 0:
        entropy_split = -((prob_class1 * math.log(prob_class1, 2)) + (prob_class2 * math.log(prob_class2, 2)))
    else:
        entropy_split = -((prob_class1 * math.log(prob_class1, 2)) + (prob_class2 * math.log(prob_class2, 2)) + (prob_class3 * math.log(prob_class3, 2)))
    return round(entropy_split,2)
```

```python
def remainder(df, left_right, label):
    total_rows = df.shape[0]
    remainder = 0
    for left_right_tree in left_right:
        if left_right_tree.shape[0] > 1:
            remainder += (left_right_tree.shape[0]/total_rows)*entropy(left_right_tree, label)
    return remainder
```

```python
def info_gain(df, attribute, label, threshold):
    #print(attribute,label,threshold)
    left_tree_df = df[df[attribute] < threshold]
    right_tree_df = df[df[attribute] > threshold]
    information_gain = entropy(df, label) - remainder(df, [left_tree_df, right_tree_df], label)
    #print(information_gain)
    return information_gain
```

```python
def select_threshold(df, attribute, label):
    values = df[attribute].tolist()
    values = [ float(x) for x in values]
    values = set(values)
    values = list(values)
    values.sort()
    max_information_gain = float("-inf")
    threshold_value = 0
    for i in range(0, len(values) - 1):
        threshold = (values[i] + values[i+1])/2
        information_gain = info_gain(df, attribute, label, threshold)
        if information_gain > max_information_gain:
            max_information_gain = information_gain
            threshold_value = threshold
    return threshold_value
```

```python
def count_class(df, label):
    class1_df = 0
    class2_df = 0
    class3_df = 0

    class1_df = df[df[label] == 1]
    class2_df = df[df[label] == 2]
    class3_df = df[df[label] == 3]
    return class1_df.shape[0], class2_df.shape[0], class3_df.shape[0]
```

```python
def choose_attribute(df, attributes, label):
    max_information_gain = float("-inf")
    best_attribute = None
    threshold = 0
    for attr in attributes:
        attribute_threshold = select_threshold(df, attr, label)
        #print(attribute_threshold)
        #print('b')
        information_gain = info_gain(df, attr, label, attribute_threshold)
        #print(information_gain)
        if information_gain > max_information_gain:
            max_information_gain = information_gain
            best_attribute = attr
            threshold = attribute_threshold
    return best_attribute, threshold
```

```python
class Node(object):
    def __init__(self, attribute, threshold):
        self.attr = attribute
        self.thres = threshold
        self.left = None
        self.right = None
        self.leaf = False
        self.predict = None
```

```python
def build_tree(df, attributes, label):
    
    class1, class2, class3 = count_class(df, label)
    #print(class1,class2,class3)
    if (class1 == 0 and class2 == 0) or (class1 == 0 and class3 == 0) or (class2 == 0 and class3 == 0):
    #if class1 == 0 or class2 == 0 or class3 == 0:
        leaf = Node(None,None)
        leaf.leaf = True
        if class1 > class2 and class1 > class3:
            leaf.predict = 1
        elif class2 > class1 and class2 > class3:
            leaf.predict = 2
        elif class3 > class1 and class3 > class2:
            leaf.predict = 3
        return leaf
    else:
        best_attribute, threshold = choose_attribute(df, attributes, label)
        tree = Node(best_attribute, threshold)
        left_tree = df[df[best_attribute] < threshold]
        right_tree = df[df[best_attribute] > threshold]
        tree.left = build_tree(left_tree, attributes, label)
        tree.right = build_tree(right_tree, attributes, label)
        return tree
    
```

```python
def predict(node, row_df):
    #if node.left == None or node.right == None:
    if node.leaf:
        return node.predict
    if row_df[node.attr] <= node.thres:
        return predict(node.left, row_df)
    elif row_df[node.attr] > node.thres:
        return predict(node.right, row_df)
```

```python
def test_predictions(root, df):
    num_data = df.shape[0]
    num_correct = 0
    prediction = []
    for index,row in df.iterrows():
        prediction_number = predict(root, row)
        prediction.append(prediction_number)
        if prediction_number == row['Class']:
            num_correct += 1
    return round(num_correct/num_data, 2),prediction
```

```python
attributes =  ['Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids',
                'Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline']
root = build_tree(X_train_copy, attributes, 'Class')
```

```python
print("Accuracy of test data")
#print(str(test_predictions(root, X_test_copy)*100.0) + '%')
result = test_predictions(root, X_test_copy)
print(str(result[0]*100.0) + '%')
#print(result[1])
```

```python
print(confusion_matrix(Y_test,result[1]))
```
